4.	Create a python program. Continue the program by creating `class Employee`. 

*Hint: Create a python class with function to determine the saalry based on the position. import this class into another python file and type the provided code below:*



When we execute the program, the output can be seen as follow.



```python
class Employee:
    def __init__(self, name, department, position):
        self.name = name
        self.department = department
        self.position = position

    def salary(self):
        if self.position == "staff":
            return "3000 USD"
        elif self.position == "manager":
            return "4000 USD"
        
s1 = Employee("Jack", "Accounting", "staff")
s2 = Employee("Mike", "R&D", "manager")

print(s1.name, ',department:', s1.department, ', position:', s1.position, ',salary:', s1.salary())
print(s2.name, ',department:', s2.department, ', position:', s2.position, ',salary:', s2.salary())
```
